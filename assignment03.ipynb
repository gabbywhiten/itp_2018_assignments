Complete the following function which checks whether a DNA sequence is valid. A valid DNA sequence will only contain **'A,a,C,c,G,g,T,t'** characters for this question. If the sequence is valid return **True**, otherwise return **False**.

*When running the block, the assert statements will raise errors if your program does not return the correct response.*


````python
def validity_check(sequence):
    nts = ['a','g','t','c','A','G','T','C']
    for x in sequence:
        if x not in nts:
            return False
    return True

assert validity_check('ACGT') == True, 'validity_check("ACGT") returns False, should return True'
assert validity_check('acgt') == True, 'validity_check("acgt") returns False, should return True'
assert validity_check('AGUC') == False, 'validity_check("AGUC") returns True, should return False'
assert validity_check('PPPP') == False, 'validity_check("PPPP") returns True, should return False'
````

#### 1.2)
Complete the following function which will return the reverse complement of a DNA sequence.

The reverse complement is calculated by reversing the sequence and substituting: 
+ 'A' with 'T'
+ 'T' with 'A'
+ 'G' with 'C'
+ 'C' with 'G'

EG:
```
reverse_complement('AATC')

step 1 - reverse the sequence:
    reverse = 'CTAA'
    
step 2 - replace the characters with their complement:
    'C' -> 'G'
    'T' -> 'A'
    'A' -> 'T'
    'A' -> 'T'
    
    complement = 'GATT' 
```

*When running the block, the assert statements will raise errors if your program does not return the correct response.*

````python
def reverse_complement(sequence):
    #Reverse the sequence given.
    reverse_sequence = sequence[::-1]
    #Replace that sequence with the complementary letters.
    reverse_complement_seq = []
    for x in reverse_sequence:
        if x in ('C'):
            reverse_complement_seq.append('G')
        elif x in ('c'):
            reverse_complement_seq.append('g')
        elif x in ('A'):
            reverse_complement_seq.append('T')
        elif x in ('a'):
            reverse_complement_seq.append('t')
        elif x in ('G'):
            reverse_complement_seq.append('C')
        elif x in ('g'):
            reverse_complement_seq.append('c')
        elif x in ('T'):
            reverse_complement_seq.append('A')
        elif x in ('t'):
            reverse_complement_seq.append('a')
        else:
            print('Error, enter valid sequence.')
    reverse_complement_sequence = "".join(reverse_complement_seq)
    print('The original sequence was {} and the reverse complement is {}.'.format(sequence, reverse_complement_sequence))
    return reverse_complement_sequence

assert reverse_complement('AAGCT') == 'AGCTT', 'reverse_complement("AAGCT") should return "AGCTT"'
assert reverse_complement('tggca') in ['tgcca','TGCCA'], 'reverse_complement("tggca") should return "tgcca"'
````

Complete the following function which takes a string and an integer which represents the shift and returns an encoded string.

````python
#Get message from user.
def get_message():
    print('Enter message to be encoded.')
    return input()

#Get the shift from user.

def get_shift():
    shift = 0
    while True:
        print('Enter shift to encode message.')
        shift = int(input())
        #Check to make sure shift is less than or equal to 26.
        if shift >= 1 and shift <= 26:
            return shift
def get_crypto_message(message, shift):
    translated = ''
    for symbol in message:
        if symbol.isalpha():
            num = ord(symbol)
            num += shift
            if symbol.isupper():
                if num > ord('Z'):
                    num -= 26
                elif num < ord('A'):
                    num += 26
            elif symbol.islower():
                if num > ord('z'):
                    num -= 26
                elif num < ord('a'):
                    num += 26
            translated += chr(num)
        else:
            translated += symbol
    return translated
            
#message = get_message()
#shift = get_shift()

print('Your translated text is:')
print(get_crypto_message(message, shift))
        
#assert encode_string('ABCD', 1) == 'BCDE', "encode_string('ABCD', 1) should return 'BCDE'"
#assert encode_string('FOOBAR', 3) == 'IRREDU', "encode_string('FOOBAR', 3) should return 'IRREDU'"
#assert encode_string('XYZ') == 'CDE', "encode_string('XYZ', 3) should return 'CDE'"
````

````python
#Get message from user.
def get_message():
    print('Enter message to be decoded.')
    return input()

#Get the shift from user.

def get_shift():
    shift = 0
    while True:
        print('Enter shift to decode message.')
        shift = int(input())
        #Check to make sure shift is less than or equal to 26.
        if shift >= 1 and shift <= 26:
            return shift
def get_decrypto_message(message, shift):
    shift = -shift
    translated = ''
    for symbol in message:
        if symbol.isalpha():
            num = ord(symbol)
            num += shift
            if symbol.isupper():
                if num > ord('Z'):
                    num -= 26
                elif num < ord('A'):
                    num += 26
            elif symbol.islower():
                if num > ord('z'):
                    num -= 26
                elif num < ord('a'):
                    num += 26
            translated += chr(num)
        else:
            translated += symbol
    return translated
            
#message = get_message()
#shift = get_shift()

print('Your translated text is:')
print(get_decrypto_message(message, shift))

assert get_decrypto_message('MNO', 10) == 'CDE', "decode_string('MNO', 10) should return 'CDE'"
#assert get_crypto_message('PYTHON') == 'KTOCJI', "decode_string('PYTHON') should return 'KTOCJI'"

assert get_decrypto_message(get_crypto_message('ABCD', 1), 1) == 'ABCD', "decode_string(encode_string('ABCD', 1), 1) should return 'ABCD'"
assert get_decrypto_message(get_crypto_message('FOOBAR', 3), 3) == 'FOOBAR', "decode_string(encode_string('FOOBAR', 3), 3) should return 'FOOBAR'"
#assert get_decrypto_message(get_crypto_message('XYZ')) == 'XYZ', "decode_string(encode_string('XYZ')) should return 'XYZ'"
````
